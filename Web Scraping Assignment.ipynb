```python
import requests
website_url=requests.get("https://en.wikipedia.org/wiki/List_of_states_in_India_by_past_population").text
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(website_url,"lxml")
print(soup.prettify())
```

```python
my_table=soup.find('table',{'class':'wikitable sortable'})
my_table
```

```python
links=my_table.findAll('a')
links
```

```python
states = []
for link in links:
     states.append(link.get('title'))
print(states) 
```
